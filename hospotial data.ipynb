```python
import numpy as np

cholesterol = np.array([180, 195, 210, 225, 240, 260, 200, 215, 230, 245])
blood_pressure = np.array([118, 122, 130, 135, 140, 145, 125, 132, 138, 142])

np.cov(cholesterol, blood_pressure, ddof=1)
```

```python
np.corrcoef(cholesterol, blood_pressure)
```

```python
import numpy as np
cholesterol = np.array([180, 195, 210, 225, 240, 260, 200, 215, 230, 245])
import matplotlib.pyplot as plt
plt.hist(cholesterol, bins=7, edgecolor='black', color='red', density=False)
plt.xlabel('Cholesterol Levels')
plt.ylabel('Frequency')
plt.title('Distribution of Cholesterol Levels')
plt.show()
```
